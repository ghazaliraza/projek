Matrix Inversion

1. buat matrix X ukuran 2x2 menggunakan numpy

```python
# Membuat matriks X ukuran 2x2
X = np.array([[1, 2],
              [3, 4]])

print(X)
```

```python
# Matriks X
X = np.array([[1, 2],
              [3, 4]])

# Invers matriks X
Xinv = np.linalg.inv(X)

print(Xinv)
```

```python
# Matriks X
X = np.array([[1, 2],
              [3, 4]])

# Invers X
Xinv = np.linalg.inv(X)

# Perkalian X dengan Xinv
hasil = np.dot(X, Xinv)

print(hasil)
```

2. selesaikan persamaan berikut menggunakan numpy, tensorflow dan pytorch

```python
#numpy

A = np.array([[4, 2],
              [5, -3]])

B = np.array([[4],
              [-7]])

X = np.linalg.inv(A) @ B
print(X)
```

```python
import torch
```

```python
#pytorch

A = torch.tensor([[4., 2.],
                  [5., -3.]])

B = torch.tensor([[4.],
                  [-7.]])

X = torch.inverse(A) @ B
print(X)
```

```python
import tensorflow as tf
```

```python
#tensorflow
A = tf.constant([[4., 2.],
                 [5., -3.]])

B = tf.constant([[4.],
                 [-7.]])

X = tf.linalg.inv(A) @ B
print(X)
```
