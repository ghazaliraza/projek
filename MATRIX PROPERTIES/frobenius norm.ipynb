```
import numpy as np
import torch
import tensorflow as tf
```

```
#menggunakan numpy, pytoch dan tensorflow buatlah matrix ukuran 3x3 lalu hitung nilai frobenius norm
```

```
A = np.array([[1, 1, 3],
              [1, 3, 1],
              [3, 1, 1]])

frob_A = np.linalg.norm(A)
print("Frobenius Norm:", frob_A)
```

```
A_pt = torch.tensor([[1, 1, 3],
                    [1, 3, 1],
                    [3, 1, 1]],dtype=torch.float32)

frob_pt = torch.linalg.norm(A_pt)
print("Frobenius Norm:", frob_pt)
```

```
A_tf = tf.Variable([[1, 1, 3],
                    [1, 3, 1],
                    [3, 1, 1]], dtype=tf.float32)

frob_tf = tf.norm(A_tf)
print("Frobenius Norm:", frob_tf)
```
