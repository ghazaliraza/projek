24. Matrix Multiplication

soal 1

```python
import numpy as np
```

```python
A = np.array([[1, 3, 7],
              [4, 1, 9]])
B = np.array([[2],
              [4],
              [6]])

print("Matrix A:")
print(A)

print("\nVector B:")
print(B)
```

```python
import torch

A_pt = torch.tensor([[1, 3, 7],
                     [4, 1, 9]])

B_pt = torch.tensor([[2],
                     [4],
                     [6]])
print("Matrix A_pt:")
print(A_pt)

print("\nVector B_pt:")
print(B_pt)
```

```python
import tensorflow as tf
A_tf = tf.constant([[1, 3, 7],
                    [4, 1, 9]])

B_tf = tf.constant([[2],
                    [4],
                    [6]])

print("Matrix A_tf:")
print(A_tf)

print("\nVector b_tf:")
print(B_tf)
```

soal 3

```python
import numpy as np

A = np.array([[1, 3, 7],
              [4, 1, 9]])

B = np.array([[2],
              [4],
              [6]])
np.dot(A, B)
```

```python
import torch

A_pt = torch.tensor([[1, 3, 7],
                  [4, 1, 9]])

B_pt = torch.tensor([[2],
                  [4],
                  [6]])
torch.matmul(A_pt, B_pt)
```

```python
import tensorflow as tf

A_tf = tf.constant([[1, 3, 7],
                    [4, 1, 9]])

B_tf = tf.constant([[2],
                    [4],
                    [6]])
tf.matmul(A_tf, B_tf)
```

```python
import numpy as np

A = np.array([[1, 3, 7],
              [4, 1, 9]])

B = np.array([[1, 3, 7],
              [4, 1, 9],
              [6, 5, 2]])

np.dot(A, B)
```

```python
import torch

A_pt = torch.tensor([[1, 3, 7],
                    [4, 1, 9]])

B_pt = torch.tensor([[1, 3, 7],
                    [4, 1, 9],
                    [6, 5, 2]])

torch.matmul(A_pt, B_pt)
```

```python
import tensorflow as tf

A_tf = tf.constant([[1, 3, 7],
                    [4, 1, 9]])

B_tf = tf.constant([[1, 3, 7],
                    [4, 1, 9],
                    [6, 5, 2]])

tf.matmul(A_tf, B_tf)
```
