# *SYMMETRIC AND IDENTITY MATRIKS*

```python
import numpy as np
```

```python
X_sym = np.array([[2, 5, 5], [0, 6, 1], [0, 0, 5]])
X_sym
```

Baris ini membuat sebuah matriks $3 \times 3$ bernama X_sym.

```python
X_sym.T
```

Simbol .T digunakan untuk melakukan transpose, yaitu mengubah baris menjadi kolom dan kolom menjadi baris.

```python
X_sym.T == X_sym
```

```python
import torch
```

 Ini adalah fungsi dari library pytorch yang telah di impor sebagai torch. Fungsi ini digunakan untuk membuat sebuah tensor.

```python
I = torch.tensor([[2, 5, 5], [0, 6, 1], [0, 0, 5]])
I
```

 Variabel I di sini merepresentasikan matriks identitas ($I$).

```python
x_pt = torch.tensor([2, 5, 5])
x_pt
```

Ini adalah sebuah vektor kolom tensor 1D yang berisi tiga angka: 25, 2, dan 5.

```python
torch.matmul(I, x_pt)
```

matmul adalah singkatan dari matrix multiplication. Baris ini mengalikan Matriks Identitas (I) dengan vektor $x_{pt}$.Hasilnya: Sesuai hukum aljabar linear, setiap matriks atau vektor yang dikalikan dengan Matriks Identitas akan menghasilkan dirinya sendiri ($Ix = x$). Jadi, output dari kodingan ini tetaplah [25, 2, 5].
