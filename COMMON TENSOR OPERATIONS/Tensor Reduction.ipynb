soal 3

```python
import numpy as np
```

```python
X = np.array([[1, 7, 3],
              [4, 5, 6]])
X = np.sum(X, axis=1)
print(X)
```

```python
import torch
```

```python
X_pt = torch.tensor([[1, 7, 3],
                     [4, 5, 6]])

X = torch.sum(X_pt, dim=1)
print(X)
```

```python
import tensorflow as tf
```

```python
X_tf = tf.constant([[1, 7, 3],
                    [4, 5, 6]])

X_tf = tf.reduce_sum(X_tf, axis=1)
print(X_tf)
```

```python
import numpy as np

X = np.array([[1, 7, 3],
              [4, 5, 6]])

X_min = np.min(X)
print(X_min)
```

```python
import torch

X_pt = torch.tensor([[1, 7, 3],
                     [4, 5, 6]])

X_pt_min = torch.min(X_pt)
print(X_pt_min)
```

```python
import tensorflow as tf

X_tf = tf.constant([[1, 7, 3],
                    [4, 5, 6]])

X_tf_min = tf.reduce_min(X_tf)
print(X_tf_min)
```

```python
import numpy as np

X = np.array([[1, 7, 3],
              [4, 5, 6]])

X_max = np.max(X)
print(X_max)
```

```python
import numpy as np

X = np.array([[1, 7, 3],
              [4, 5, 6]])

rata_rata = np.mean(X)
print(rata_rata)
```

```python
import torch

X_pt = torch.tensor([[1, 7, 3],
                     [4, 5, 6]], dtype=torch.float32)

rata_rata = torch.mean(X_pt)
print(rata_rata)
```

```python
import tensorflow as tf

X_tf = tf.constant([[1, 7, 3],
                    [4, 5, 6]], dtype=tf.float32)

rata_rata = tf.reduce_mean(X_tf)
print(rata_rata)
```

```python
import numpy as np

X = np.array([[1, 7, 3],
              [4, 5, 6]])

X = np.prod(X)
print(X)
```

```python
import torch

X_pt = torch.tensor([[1, 7, 3],
                     [4, 5, 6]])

X_pt = torch.prod(X_pt)
print(X_pt)
```

```python
import tensorflow as tf

X_tf = tf.constant([[1, 7, 3],
                    [4, 5, 6]])

X_tf = tf.reduce_prod(X_tf)
print(X_tf)
```
