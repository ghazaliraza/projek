```
import numpy as np
```

```
import torch
```

```
import tensorflow as tf
```

```
#buat vector x dan y dengan ukuran 1x3 menggunakan numpy
x  = np.array([1, 3, 1])
y  = np.array([3, 1, 1])
dot_np = np.dot(x, y)
print( x, y, " dot:", dot_np)

```

```
#buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch
x_pt   = torch.tensor([1, 3, 1])
y_pt   = torch.tensor([3, 1, 1])
dot_pt = torch.dot(x_pt, y_pt)
print( x_pt, y_pt, " dot:", dot_pt)
```

```
#buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow
x_tf   = tf.Variable([1, 3, 1])
y_tf   = tf.Variable([3, 1, 1])
dot_tf = tf.reduce_sum(tf.multiply(x_tf, y_tf))
print( x_tf, y_tf, " dot:", dot_tf)
```

```
# y transpose
Y = np.array([[42, 4, 7, 99],
              [-99, -3, 17, 22]])
Y_T = Y.T
print("Y^t : " , Y_T)
```

```
# Hadamard product(elemen kali elemen)
A = np.array([[25, 10],
              [-2,  1]])

B = np.array([[-1, 7],
              [10, 8]])

hadamard = A * B
print("hadamard: ", hadamard)
```

```
#dot product tensor w dan x
w = np.array([-1,  2, -2])
x = np.array([ 5, 10,  0])

dot_wx = np.dot(w, x)
print("dot wx =", dot_wx)
```
