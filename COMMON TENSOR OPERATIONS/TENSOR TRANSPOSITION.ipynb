# 1. buat matrix X dengan ukuran 2x3 menggunakan numpy

```python
import numpy as np
```

```python
X = np.array([[25, 1, 5], [0, 61, 60]])
X
```

# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch

```python
import torch as pt
```

```python
X_pt = pt.tensor([[25, 1, 5], [0, 61, 60]])
X_pt
```

# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow

```python
import tensorflow as tf
```

```python
X_tf = tf.Variable ([[25, 1, 5], [0, 61, 60]])
X_tf
```

# 2. lakukan transpose pada matrix X, X_pt dan X_tf

```python
# transpose numpy
X.T
```

```python
#transpose pytorch
X_pt.T
```

```python
#transpose tensorflow
tf.transpose(X_tf)
```
