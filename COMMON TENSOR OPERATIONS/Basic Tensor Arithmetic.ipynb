2. buat matrix Y dengan ukuran 2x3 menggunakan numpy,pytorch,dan tensorflow

```python
# Matrix Y ukuran 2x3 numpy
Y = np.array([
    [7, 8, 9],
    [10, 11, 12]
])

print("Matrix Y (NumPy):")
print(Y)
```

```python
# Matriks Y ukuran 2x3 pytorch

Y_pt = torch.tensor([
    [7, 8, 9],
    [10, 11, 12]
])

print("Matrix Y (PyTorch):")
print(Y_pt)
```

```python
# Matriks Y ukuran 2x3 tensorflow

Y_tf = tf.constant([
    [7, 8, 9],
    [10, 11, 12]
])

print("Matrix Y (TensorFlow):")
print(Y_tf)
```

kalikan matrix X dengan Y,Y_pt,Y_tf

```python
#Y

X = np.array([[1, 2, 3],
              [4, 5, 6]])

Y = np.array([[7, 8, 9],
              [10, 11, 12]])

hasil_np = X * Y
print(hasil_np)
```

```python
#Y_pt

X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]])

Y_pt = torch.tensor([[7, 8, 9],
                     [10, 11, 12]])

hasil_pt = X_pt * Y_pt
print(hasil_pt)
```

```python
#Y_tf

X_tf = tf.constant([[1, 2, 3],
                    [4, 5, 6]])

Y_tf = tf.constant([[7, 8, 9],
                    [10, 11, 12]])

hasil_tf = X_tf * Y_tf
print(hasil_tf)
```
