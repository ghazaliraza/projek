# Basis, Orthogonal, and Orthonormal Vectors.

```python
import numpy as np
```

Baris ini bertujuan untuk memanggil library NumPy ke dalam program.

```python
i = np.array([0, 5])
i
```

 vektor pertama bernama i yang memiliki koordinat x=0 dan y=5.

dan Vektor ini mengarah lurus ke atas pada sumbu y sepanjang 5 satuan


```python
j = np.array([0, 5])
j
```

 vektor kedua bernama j yang nilainya sama persis dengan vektor I (identitas).
​
Karena nilainya sama, kedua vektor ini saling berhimpit atau sejajar, bukan tegak lurus.

```python
np.dot(i, j)
```

 Fungsi ini menghitung perkalian elemen yang sejajar (0x0)+(5x5)=25

Kesimpulannya kodingan ini membuktikan bahwa vektor 0,5 tidak orthogonal karena hasil np.dot nya adalah 25.sedangkan untuk syarat orthogonal hasil dot productnya itu harus 0
